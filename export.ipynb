```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
from supabase import create_client
```

```python
export_path = 'processed/export.csv'
env_path = f'.env'
load_dotenv(dotenv_path=env_path)
```

```python
db_url = os.getenv('SUPABASE_URL')
db_service_role = os.getenv('SUPABASE_SERVICE_ROLE')
supabase = create_client(db_url, db_service_role)
```

```python
data = pd.read_csv(filepath_or_buffer=export_path,index_col=0)
data.head()
```

```python
table = data.to_dict(orient='records')
response = supabase \
    .table('coach_year') \
    .upsert(table). \
    execute()
if response.data:
    print("Data inserted successfully!")
else:
    print("Error:", response.error)
```
