In order to remain on the supabase free trial in the NFL offseason, the database will receive these dummy updates to maintain activity.

```python
import os
import pandas as pd
from supabase import create_client
from dotenv import load_dotenv
import datetime
```

```python
env_path = f'.env'
load_dotenv(dotenv_path=env_path)
```

```python
db_url = os.getenv('SUPABASE_URL')
db_service_role = os.getenv('SUPABASE_SERVICE_ROLE')
supabase = create_client(db_url, db_service_role)
```

```python
currentTime = datetime.datetime.now(datetime.timezone.utc).isoformat()
timeDf = pd.DataFrame([['time',currentTime]],columns=['entry','value'])
data = timeDf.to_dict(orient='records')
response = supabase.table('update_time').upsert(data).execute()
if response.data:
    print("Data inserted successfully!")
else:
    print("Error:", response.error)
# timeDf.to_sql(name='update_time',con=engine,if_exists='replace',index=False)
```
