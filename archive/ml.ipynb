### Without cleaning the data

I crudely called every coach as "fired" if it was their last year with a team.  Meaning, coaches who retired are incorrectly labeled as fired.  This is likely particularly damaging to our prediction model as those coaches who retire are likely to have some of the best resumes.
However, I believe it is important to use this uncleaned data as a baseline for an accuracy score to beat.

```python
# Initialize the model
model = LogisticRegression()

# Train the model
model.fit(X_train_scaled, y_train)

# Predict
y_pred = model.predict(X_test_scaled)

# Evaluate
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(classification_report(y_test, y_pred))
```
